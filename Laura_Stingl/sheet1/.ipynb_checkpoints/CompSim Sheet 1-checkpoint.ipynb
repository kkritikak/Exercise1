### e)(1/1)

```python
# generates n random numbers between 0 and 1
def congr_gen_nums(seed, n):
    a = 1664525
    c = 1013904223
    m = 2**32
    
    x = np.zeros(n)
    
    x[0] = (a * seed + c) % m
    
    for i in range(n - 1):
        x[i + 1] = (a * x[i] + c) % m
    
    return x/m
```

```python
rand_x_s = congr_gen_nums(302, N)
rand_y_s = congr_gen_nums(302, N)

plt.scatter(rand_x, rand_y, alpha=0.01)
```
